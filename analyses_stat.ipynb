```python
import pandas as pd

# Charger le fichier CSV (à adapter si nécessaire)
input_file = "./data/irasshai_sales.csv"
sales_df = pd.read_csv(input_file, encoding="utf-8")

print("Nombre de lignes :", len(sales_df))
sales_df.head()
```

```python
output_file = "./data/irasshai_sales_data.csv"
sales_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"Fichier exporté vers : {output_file}")
```

```python
print(f"Nombre total de commandes : {len(sales_df)}")
print(f"Chiffre d'affaires total : {sales_df['Sales'].sum():.2f} €")
```

```python
print("Statistiques par catégorie :")

category_stats = sales_df.groupby('Category').agg({
    'Sales': 'sum',
    'Units Sold': 'sum',
    'Order ID': 'count'
}).round(2)

category_stats.columns = ['CA Total (€)', 'Unités vendues', 'Nb commandes']
display(category_stats.sort_values('CA Total (€)', ascending=False))
```

```python
print("\n Top 5 des régions par CA :")

region_stats = sales_df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(5)

for region, sales in region_stats.items():
    print(f"  {region}: {sales:.2f} €")
```

```python
print("\n Top 10 des produits les plus vendus :")

top_products = sales_df.groupby('Product').agg({
    'Units Sold': 'sum',
    'Sales': 'sum'
}).sort_values('Sales', ascending=False).head(10)

for product, row in top_products.iterrows():
    print(f"  {product}: {row['Units Sold']} unités - {row['Sales']:.2f} €")
```
